```python
import os
import sys
import urllib.request
import pandas as pd
import json #파싱하기위해 json 패키지 소환!
import re

client_id = "wNLpq4kfjLjPc_dxMRw_"
client_secret = "zYYJnd3qIt"
encText = urllib.parse.quote(input("무엇을 검색하시겠습니까?  : "))

#1~500 까지 검색해볼거야! start , end , 한번에 최대 100개!

cnt = 0  #몇개 검색했는지 측정!
start = 1
end = 500 
display =100

df = pd.DataFrame(columns=('title','link','description'))

# i를 반복할꺼야, 시작 start , end , display) -> 1~500 , 100간격으로!
for i in range(start,end,display):
  #url = "https://openapi.naver.com/v1/search/webkr?query=" + encText # 웹문서를 검색할꺼야! 1~100 , 1~100, 1~100  -> 1~100 , 101~200 , 201~300
  url = "https://openapi.naver.com/v1/search/webkr?query=" + encText + "&display=" + str(display) + "&start=" + str(i)  #한번에 100건씩! , 시작위치는 우리가 바꿔줄꺼야!
  request = urllib.request.Request(url)
  request.add_header("X-Naver-Client-Id",client_id)
  request.add_header("X-Naver-Client-Secret",client_secret)
  response = urllib.request.urlopen(request)
  rescode = response.getcode()
  if(rescode==200):
     response_body = response.read()
     response_dict =  json.loads(response_body.decode('utf-8'))  #요청해서 응답받은 내용을 데이터딕셔너리 형태로 저장!

     items = response_dict['items'] #불필요한 내용 다 지우고! items에 저장된 데이터만 추려와 items에 저장!

      #추려왔으면 우리가 만든 df라는 데이터프레임에 하나하나 넣어줘야함! 0번째 줄부터 items에 있는 데이터 만큼! 채워줄꺼야!

     for it in range(0,len(items)):
       rm_tag = re.compile('<.*?>')  #지울 테그들을 정규표현식으로 탐색할꺼야! re.compile('지워줄내용') -> rm_tag에 지워줄 내용들 저장!   <b> , </b>
       title = re.sub(rm_tag,'', items[it]['title']) #sub은 문자 대체!  앞에 있는 애를 뒤에있는애로 대체!
       link = items[it]['link']
       description = re.sub(rm_tag, '', items[it]['description'])
       df.loc[cnt] = [title, link, description]
       cnt+=1

  else:
      print("Error Code:" + rescode)


df
```
